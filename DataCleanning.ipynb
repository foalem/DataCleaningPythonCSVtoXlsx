```python
import texthero as hero
import pandas as pd
```

```python
news_content = df[["Answer"]]
news_content.head()
```

```python
print(news_content)
```

```python
news_content['Answer'].iloc[0]
```

```python
#create custom pipeline
from texthero import preprocessing

custom_pipeline = [preprocessing.fillna,
                   preprocessing.lowercase,
                   preprocessing.remove_urls,
                   preprocessing.remove_html_tags,
                   preprocessing.remove_urls,
                   preprocessing.remove_diacritics
                   ]
```

```python
news_content['clean_custom_Answer'] = news_content['Answer'].pipe(hero.clean, custom_pipeline)
```

```python
print(news_content['clean_custom_Answer'])
```

```python
news_content['clean_custom_Answer'].iloc[0]
```

```python
df['Answer'] = news_content['clean_custom_Answer']
```
